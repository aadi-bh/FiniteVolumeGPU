```python
%matplotlib notebook

from matplotlib import rc

import numpy as np
from matplotlib import pyplot as plt
```

```python
def desingularize(x, eps):
    return np.sign(x)*np.maximum(np.minimum(x*x/(np.float32(2.0)*eps)+np.float32(0.5)*eps, eps), abs(x))

eps = np.float32(2.0e-22)
x = np.linspace(-1.5*eps, 1.5*eps, 500).astype(np.float32)
y = desingularize(x, eps)

plt.figure()
plt.plot(x, x, label='orig')
plt.plot(x, y, label='limited')
plt.xlabel('x')
plt.ylabel('Desingularized')
plt.legend()
```
